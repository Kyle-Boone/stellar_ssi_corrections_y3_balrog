```python
import sys
sys.path.insert(1, '/afs/hep.wisc.edu/home/kkboone/software/StarWeights/FinalPipeline')

import fitsio
import numpy as np
import healpy as hp
import healsparse as hsp
import skyproj
import matplotlib.pyplot as plt
import Config
import StellarConfig as strConfig

import matplotlib.style
import matplotlib
matplotlib.style.use('des_dr1')
```

```python
origCondFiles = strConfig.origCondFiles
stelFile = strConfig.stelFile
conditions = strConfig.conditions

nside_coverage = 32
```

# A Few Survey Property Maps:

Do stellar density, fwhm (r), exposure time (i), and skybrite (z)

### Stellar Density

```python
stelDens = fitsio.read(stelFile)['I'].flatten() # Loads in the stellar density.
stelDens[np.where(stelDens < -100)] = hp.UNSEEN

hspStelDens = hsp.HealSparseMap(nside_coverage=nside_coverage, healpix_map=stelDens)

fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111)
sp = skyproj.DESSkyproj(ax=ax)
sp.draw_hspmap(hspStelDens, cmap = 'viridis')
plt.clim(0, 8)
sp.draw_inset_colorbar(label = 'Stellar Density' + '\n' + r'(stars/deg$^2$)', format = '%.1f')
# plt.colorbar(location = 'bottom', label = 'Stars per Square Degree')
# fig.suptitle('Stellar Density', y = 1.0)
plt.show()
```

### FWHM Weighted Mean, r Band

```python
FWHMData = fitsio.read(origCondFiles[13]) # This reads in the data.
FWHM = np.full(12*(4096**2), hp.UNSEEN) # Gives a default value.
FWHM[FWHMData['PIXEL']] = FWHMData['SIGNAL'] # Changes all valid pixels to their corresponding signals.
FWHM[np.where(FWHM < -100)] = hp.UNSEEN # An invalid survey property marks this pixel as false.

hspFWHM = hsp.HealSparseMap(nside_coverage=nside_coverage, healpix_map=FWHM)

fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111)
sp = skyproj.DESSkyproj(ax=ax)
sp.draw_hspmap(hspFWHM, cmap = 'viridis')
plt.clim(0.8, 1.2)
sp.draw_inset_colorbar(label = 'Seeing, r Band'+'\n'+'(arcsec)', format = '%.1f')
# plt.colorbar(location = 'bottom', label = 'Arcsec')
# fig.suptitle('FWHM Weighted Mean, r Band', y = 1.0)
plt.show()
```

### Exposure Time Sum, i Band

```python
expTimeData = fitsio.read(origCondFiles[38]) # This reads in the data.
expTime = np.full(12*(4096**2), hp.UNSEEN) # Gives a default value.
expTime[expTimeData['PIXEL']] = expTimeData['SIGNAL'] # Changes all valid pixels to their corresponding signals.
expTime[np.where(expTime < -100)] = hp.UNSEEN # An invalid survey property marks this pixel as false.

hspexpTime = hsp.HealSparseMap(nside_coverage=nside_coverage, healpix_map=expTime)

fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111)
sp = skyproj.DESSkyproj(ax=ax)
sp.draw_hspmap(hspexpTime, cmap = 'viridis')
plt.clim(135, 540)
sp.draw_inset_colorbar(label = 'Exposure Time, i Band'+'\n'+'(sec)', format = '%.1f')
# plt.colorbar(location = 'bottom', label = 'Seconds')
# fig.suptitle('Exposure Time Sum, i Band', y = 1.0)
plt.show()
```

### Skybrite, z Band

```python
skyBriteData = fitsio.read(origCondFiles[59]) # This reads in the data.
skyBrite = np.full(12*(4096**2), hp.UNSEEN) # Gives a default value.
skyBrite[skyBriteData['PIXEL']] = skyBriteData['SIGNAL'] # Changes all valid pixels to their corresponding signals.
skyBrite[np.where(skyBrite < -100)] = hp.UNSEEN # An invalid survey property marks this pixel as false.

hspskyBrite = hsp.HealSparseMap(nside_coverage=nside_coverage, healpix_map=skyBrite)

fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111)
sp = skyproj.DESSkyproj(ax=ax)
sp.draw_hspmap(hspskyBrite, cmap = 'viridis')
plt.clim(6000, 10000)
sp.draw_inset_colorbar(label = 'Sky Brightness, z Band'+'\n'+'(electrons / pixel)', format = '%.1f')
# plt.colorbar(location = 'bottom', label = 'Electrons / CCD Pixel')
# fig.suptitle('Skybrite Weighted Mean, z Band', y = 1.0)
plt.show()
```

## Combined

```python
fig = plt.figure(1, figsize=(8, 6))

ax = fig.add_subplot(221)
sp = skyproj.DESSkyproj(ax=ax)
sp.draw_hspmap(hspStelDens, cmap = 'viridis')
plt.clim(0, 8)
sp.draw_inset_colorbar(label = 'Stellar Density' + '\n' + r'(stars/deg$^2$)', format = '%.1f')

ax = fig.add_subplot(222)
sp = skyproj.DESSkyproj(ax=ax)
sp.draw_hspmap(hspFWHM, cmap = 'viridis')
plt.clim(0.8, 1.2)
sp.draw_inset_colorbar(label = 'Seeing, r Band'+'\n'+'(arcsec)', format = '%.1f')

ax = fig.add_subplot(223)
sp = skyproj.DESSkyproj(ax=ax)
sp.draw_hspmap(hspexpTime, cmap = 'viridis')
plt.clim(135, 540)
sp.draw_inset_colorbar(label = 'Exposure Time, i Band'+'\n'+'(sec)', format = '%.1f')

ax = fig.add_subplot(224)
sp = skyproj.DESSkyproj(ax=ax)
sp.draw_hspmap(hspskyBrite, cmap = 'viridis')
plt.clim(6000, 10000)
sp.draw_inset_colorbar(label = 'Sky Brightness, z Band'+'\n'+'(electrons / pixel)', format = '%.1f')

plt.show()
```
