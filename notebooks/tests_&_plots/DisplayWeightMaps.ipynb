## Limit on Cycles

```python
weights = []

stellarDir = '/hdfs/bechtol/balrog/y3/y3a2_survey_conditions_maps/Kyle_Stuff/MaximumLikelihood/FinalPipeline/Phoenix_Class_1.5/'
weightFile = stellarDir + 'Effective_Weights.fits'
pix = fitsio.read(weightFile)['PIX']
weights.append(fitsio.read(weightFile)['WEIGHT'])

stellarDir = '/hdfs/bechtol/balrog/y3/y3a2_survey_conditions_maps/Kyle_Stuff/MaximumLikelihood/FinalPipeline/OLD_Phoenix_Class_1.5/'
weightFile = stellarDir + 'Effective_Weights.fits'
weights.append(fitsio.read(weightFile)['WEIGHT'])
```

```python
fullRatios = []

for i in np.arange(2):
    fullRatios.append(np.full(12*(512**2), hp.UNSEEN))
    fullRatios[-1][pix] = weights[i]
    
fullRatios = np.array(fullRatios)
```

```python
ratioComp = np.full(12*(512**2), hp.UNSEEN)
ratioComp[pix] = weights[0] - weights[1]
```

```python
nside_coverage = 32

fullHSP = []

for i in np.arange(2):
    fullHSP.append(hsp.HealSparseMap(nside_coverage=nside_coverage, healpix_map=fullRatios[i]))
    
compHSP = hsp.HealSparseMap(nside_coverage=nside_coverage, healpix_map=ratioComp)
```

```python
# The default DES projection is a McBrydeSkymap.
fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111)
sp = skyproj.DESSkyproj(ax=ax)
sp.draw_hspmap(fullHSP[0], cmap = 'viridis')
plt.clim(0.67, 1.5)
plt.colorbar(location = 'right', label = 'Effective Weight', fraction = 0.0267)
fig.suptitle(r'Limit Cycle Training, $\sigma$ = ' + "{:.3f}".format(np.std(weights[0])), y = 0.9)
plt.show()
```

```python
# The default DES projection is a McBrydeSkymap.
fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111)
sp = skyproj.DESSkyproj(ax=ax)
sp.draw_hspmap(fullHSP[1], cmap = 'viridis')
plt.clim(0.67, 1.5)
plt.colorbar(location = 'right', label = 'Effective Weight', fraction = 0.0267)
fig.suptitle(r'No Limit Training, $\sigma$ = ' + "{:.3f}".format(np.std(weights[1])), y = 0.9)
plt.show()
```

```python
# The default DES projection is a McBrydeSkymap.
fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111)
sp = skyproj.DESSkyproj(ax=ax)
sp.draw_hspmap(compHSP, cmap = 'bwr')
plt.clim(-0.01, 0.01)
plt.colorbar(location = 'right', label = 'Effective Weight', fraction = 0.0267)
fig.suptitle(r'Residuals of Runs, $\sigma$ = ' + "{:.3f}".format(np.std(ratioComp[pix])), y = 0.9)
plt.show()
```

## Varied Percentages

```python
pers = np.linspace(10, 50, 9)
seeds = np.array([0, 1])
```

```python
weights = []

for i in np.arange(len(seeds)):
    seedWeights = []
    for j in np.arange(len(pers)):
        stellarDir = '/hdfs/bechtol/balrog/y3/y3a2_survey_conditions_maps/Kyle_Stuff/MaximumLikelihood/FinalPipeline/Tests/Percent_Used/' + str(int(pers[j])) + '/'

        weightFile = stellarDir + 'Effective_Weights_'+str(seeds[i])+'.fits'
        if i == 0:
            if j == 0:
                pix = fitsio.read(weightFile)['PIX']
        seedWeights.append(fitsio.read(weightFile)['WEIGHT'])
    weights.append(np.array(seedWeights))
weights = np.array(weights)
```

```python
fullRatios = []

for i in np.arange(len(seeds)):
    seedRatios = []
    for j in np.arange(len(pers)):
        seedRatios.append(np.full(12*(512**2), hp.UNSEEN))
        seedRatios[-1][pix] = weights[i][j]
    fullRatios.append(np.array(seedRatios))
    
fullRatios = np.array(fullRatios)
```

```python
fullComps = []

for i in np.arange(len(pers)):
    comp = np.full(12*(512**2), hp.UNSEEN)
    comp[pix] = weights[0][i] - weights[1][i]
    fullComps.append(comp)

fullComps = np.array(fullComps)
```

```python
nside_coverage = 32

fullHSP = []

for i in np.arange(len(seeds)):
    seedHSP = []
    for j in np.arange(len(pers)):
        seedHSP.append(hsp.HealSparseMap(nside_coverage=nside_coverage, healpix_map=fullRatios[i][j]))
    fullHSP.append(seedHSP)
    
compHSP = []
for i in np.arange(len(fullComps)):
    compHSP.append(hsp.HealSparseMap(nside_coverage=nside_coverage, healpix_map=fullComps[i]))
```

## Potential Fits

```python
import scipy as sc
```

```python
def exp(x, C, x0, k):
    return (C*((x - x0)**(-1 * k)))
```

```python
popt, pcov = sc.optimize.curve_fit(exp, pers, residStds, bounds = ([0.1, 1., 0.], [10., 10., 2.]))
```

```python
popt
```

```python
perr = np.sqrt(np.diag(pcov))
perr
```

```python
cmap = plt.get_cmap("tab10")
plt.figure(dpi = 120)
plt.plot(np.linspace(9, 100, 500), exp(np.linspace(9, 100, 500), *popt), color = cmap(0), label = 'Fit')
plt.scatter(pers, residStds, color = cmap(1), zorder = 3, label = 'Calculated')
plt.legend(loc = 'lower left')
plt.ylim([0, 0.40])
plt.xlim([0, 60])
plt.text(30, 0.2, r'$y = C(x-x_0)^{-k}$' '\n' r'$k = 0.45\pm 0.01$' '\n' r'$C = 0.46\pm 0.02$' '\n' r'$x_0 = 8.0\pm 0.2$', bbox={'facecolor': 'white', 'alpha': 1, 'pad': 10}, zorder = 3)
plt.ylabel(r'$\sigma$ of Effective Weight Map Residuals')
plt.xlabel('Percent of Balrog Objects Used')
plt.title('Residual Improvements with Object Counts')
plt.grid(zorder = 0)
plt.show()
```
